```python
import numpy as np
import matplotlib.pyplot as plt
from math import cos, sin, pi

# Исходный объект
A = np.array([
    [-100, -100, 1],
    [-100, 100, 1],
    [100, 100, 1],
    [0, 0, 1],
    [100, -100, 1],
    [-100, -100, 1]
])

# 1. Уменьшение в 2 раза
scale_matrix = np.array([
    [0.5, 0, 0],
    [0, 0.5, 0],
    [0, 0, 1]
])

# 2. Поворот на 130 градусов
theta = np.radians(130)
rotate_matrix = np.array([
    [cos(theta), sin(theta), 0],
    [-sin(theta), cos(theta), 0],
    [0, 0, 1]
])

# 3. Отражение относительно y = x
reflect_matrix = np.array([
    [0, 1, 0],
    [1, 0, 0],
    [0, 0, 1]
])

# Комбинируем все преобразования:
# Последовательность применения: сначала уменьшение, потом поворот, потом отражение
transform = reflect_matrix @ rotate_matrix @ scale_matrix

# Применяем трансформацию
A_transformed = (transform @ A.T).T

# Визуализация
plt.figure(figsize=(8, 8))
plt.plot(A[:, 0], A[:, 1], label='Исходный объект', color='blue')
plt.plot(A_transformed[:, 0], A_transformed[:, 1], label='Преобразованный объект', color='red')
plt.legend()
plt.grid(True)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.gca().set_aspect('equal')
plt.title("Линейные преобразования объекта")
plt.show()
```
