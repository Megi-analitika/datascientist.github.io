```python
import numpy as np
import matplotlib.pyplot as plt
from math import cos, sin, pi

# Source object
A = np.array([
    [-100, -100, 1],
    [-100, 100, 1],
    [100, 100, 1],
    [0, 0, 1],
    [100, -100, 1],
    [-100, -100, 1]
])

# 1. Decrease by 2 times
scale_matrix = np.array([
    [0.5, 0, 0],
    [0, 0.5, 0],
    [0, 0, 1]
])

#2. 130 degree rotation
theta = np.radians(130)
rotate_matrix = np.array([
    [cos(theta), sin(theta), 0],
    [-sin(theta), cos(theta), 0],
    [0, 0, 1]
])

# 3. Reflection with respect to y = x
reflect_matrix = np.array([
    [0, 1, 0],
    [1, 0, 0],
    [0, 0, 1]
])

# Combine all the transformations
# Sequence of application: first reduce, then rotate, then reflect
transform = reflect_matrix @ rotate_matrix @ scale_matrix

# Apply the transformation
A_transformed = (transform @ A.T).T

# Visualization
plt.figure(figsize=(8, 8))
plt.plot(A[:, 0], A[:, 1], label='Source object', color='blue')
plt.plot(A_transformed[:, 0], A_transformed[:, 1], label='Transformed object', color='red')
plt.legend()
plt.grid(True)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.gca().set_aspect('equal')
plt.title("Linear transformations of the object")
plt.show()
```
